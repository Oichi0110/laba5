$$ Задание 3.1 $$
Работа с графиками и обработка массивов. Сформировать массив значений функции f(x) на заданном интервале по формуле из таблицы ниже. Вывести на экран значения аргумента и значения функции. Найти в массиве наибольшее, наименьшее, среднее значение, определить количество элементов массива, а также отсортировать его (чётные варианты – по убыванию, нечётные – по возрастанию). Построить график изменения значений функции, вывести на экран его с обозначением осей, пределов изменения функции и аргумента. На экран также вывести график прямой, значение которой равно среднему значению функции f(x). График прямой и функции оформить различными маркерами. $$ l = \sqrt[5]{|2*x - c|^3} + 0.567, при\ \ x = 12.1 \ \ и -10 <= c <= 1; \Delta c = 0.5$$

```python
x, c = 12.1, -10.0
l_array, c_array = [], []
while c <= 1:
    lds = np.power(np.abs(2 * x - c)**3, 1/5) + 0.567
    c_array.append(c)
    l_array.append(lds)
    print(f"l(c) = {lds:.3f}. При c = {c}")
    c += 0.5
print(f"Наибольшее значение для массива f(c): {np.max(l_array):.3f}\n"
      f"Наименьшее значение для массива f(c): {np.min(l_array):.3f}\n"
      f"Среднее значение для массива f(c): {np.mean(l_array):.3f}\n"
      f"Количество элементов массива: {len(l_array)}\n"
      f"Отсортированный массив по убыванию: {np.sort(l_array)[::-1]}")
plt.plot(c_array, l_array, marker='o', label='f(c)')
plt.axhline(y=np.mean(l_array), color='r', linestyle='--', label='Ср. значение f(c)')
plt.axis([-12.0, 2.0, 6.0, 10.0])
plt.xlabel("c")
plt.ylabel("f(c)")
plt.legend()
plt.title('График функции f(c) и среднего значения функции f(c)')
plt.show()
```

$$ Задание 3.2 $$ Построить графики функций, изменения и диапазон аргумента задать самостоятельно

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x, y = -5.0, -3.0
x_array, y_array, z_array = [], [], []
while x <= 4:
    if x >= 0 and y >= 0:
        lds = x**0.25 + y**0.25
        x_array.append(x)
        y_array.append(y)
        z_array.append(lds)
        print(f"z = {lds:.3f}. При x, y = {x, y}")
    x += 0.5
    y += 0.5
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_array, y_array, z_array, label='График функции z = x^0.25 + y^0.25', color='red')
plt.title('График функции z = x^0.25 + y^0.25')
plt.show()
x, y = -5.0, -3.0
x_array, y_array, z1_array = [], [], []
while x <= 4:
    lds = x**2 - y**2
    x_array.append(x)
    y_array.append(y)
    z1_array.append(lds)
    print(f"z = {lds:.3f}. При x, y = {x, y}")
    x += 0.5
    y += 0.5
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_array, y_array, z1_array, label='График функции z = x^2 - y^2', color='#ff2400')
plt.title('График функции z = x^2 - y^2')
plt.show()
x, y = -5.0, -3.0
x_array, y_array, z2_array = [], [], []
while x <= 4:
    lds = 2 * x + 3 * y
    x_array.append(x)
    y_array.append(y)
    z2_array.append(lds)
    print(f"z = {lds:.3f}. При x, y = {x, y}")
    x += 0.5
    y += 0.5
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_array, y_array, z2_array, label='График функции z = 2x + 3y', color='#50c878')
plt.title('График функции z = 2x + 3y')
plt.show()
x, y = -5.0, -3.0
x_array, y_array, z3_array = [], [], []
while x <= 4:
    lds = x**2 + y**2
    x_array.append(x)
    y_array.append(y)
    z3_array.append(lds)
    print(f"z = {lds:.3f}. При x, y = {x, y}")
    x += 0.5
    y += 0.5
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_array, y_array, z3_array, label='График функции z = x^2 + y^2', color='#000080')
plt.title('График функции z = x^2 + y^2')
plt.show()
x, y = -5.0, -3.0
x_array, y_array, z4_array = [], [], []
while x <= 4:
    lds = 2 + 2 * x + 2 * y - x**2 - y**2
    x_array.append(x)
    y_array.append(y)
    z4_array.append(lds)
    print(f"z = {lds:.3f}. При x, y = {x, y}")
    x += 0.5
    y += 0.5
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_array, y_array, z4_array, label='График функции z = 2 + 2x + 2y - x^2 - y^2', color='#ff033e')
plt.title('График функции z = 2 + 2x + 2y - x^2 - y^2')
plt.show()
```
